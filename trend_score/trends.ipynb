```python
from matplotlib import pyplot
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("data1_up_likes.csv")
fig, ax = pyplot.subplots()
df.plot(x='time', y='value', ax=ax)

model = sm.formula.ols(formula='value ~ time', data=df)
res1 = model.fit()
df.assign(fit=res1.fittedvalues).plot(x='time', y='fit', ax=ax)

df = pd.read_csv("data1_up_plays.csv")
#fig, ax = pyplot.subplots()
df.plot(x='time', y='value', ax=ax)

model = sm.formula.ols(formula='value ~ time', data=df)
res2 = model.fit()
df.assign(fit=res2.fittedvalues).plot(x='time', y='fit', ax=ax)

df = pd.read_csv("data1_up_replies.csv")
df.plot(x='time', y='value', ax=ax)

model = sm.formula.ols(formula='value ~ time', data=df)
res3 = model.fit()
df.assign(fit=res3.fittedvalues).plot(x='time', y='fit', ax=ax)

weights = [.4, .1, .5]
print("Trend Value: ", sum(map(lambda x: x[0] * x[1], (zip(weights, [res1.params[1], res2.params[1], res3.params[1]])))))
```
